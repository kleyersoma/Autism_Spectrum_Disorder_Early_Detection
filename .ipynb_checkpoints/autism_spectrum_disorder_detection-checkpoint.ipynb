The method below <code>arff2csv</code> will help us in converting the file <code>.arff</code> to a <code>.csv</code>, this method was taken from this github user <a href='https://github.com/garicchi'>garicchi</a> and his repository <a href='https://github.com/Hutdris/arff2csv'>arff2csv</a>.

```python
def arff2csv(arff_path, csv_path=None, _encoding='utf8'):
    with open(arff_path, 'r', encoding=_encoding) as fr:
        attributes = []
        if csv_path is None:
            csv_path = arff_path[:-4] + 'csv'  # *.arff -> *.csv
        write_sw = False
        with open(csv_path, 'w', encoding=_encoding) as fw:
            for line in fr.readlines():
                if write_sw:
                    fw.write(line)
                elif '@data' in line:
                    fw.write(','.join(attributes) + '\n')
                    write_sw = True
                elif '@attribute' in line:
                    attributes.append(line.split()[1])  # @attribute attribute_tag numeric
    print("Convert {} to {}.".format(arff_path, csv_path))
    return csv_path
```

<h2>Developing The Neural Network with Keras, <i>it won't be the best model for solving ASD Dataset, but why?</i></h2>
<br>
It is the time to develop the Neural Network that we are going to use for resolving the classification problem of predicting if a child has ASD, the kind of Neural Network we are going to develop is called <b>Multilayer Perceptron</b>, it is a simple Neural Network, in another notebook we are going to use a state of the art architecture, in the post <a href='https://towardsdatascience.com/multi-layer-neural-networks-with-sigmoid-function-deep-learning-for-rookies-2-bf464f09eb7f'>Multi-Layer Neural Networks with Sigmoid Function</a> it is well explained how a Multilayer Perceptron works and learns.

Let's define a function called <code>input_dim()</code> that takes as parameter the length of columns from the dataset or <i>training set</i>.

```python
def input_dim(x=X_train.shape[1]):
    return x
```

```python

def f1_weighted(true, pred): #shapes (batch, 4)

    #for metrics include these two lines, for loss, don't include them
    #these are meant to round 'pred' to exactly zeros and ones
    #predLabels = K.argmax(pred, axis=-1)
    #pred = K.one_hot(predLabels, 4) 


    ground_positives = K.sum(true, axis=0)       # = TP + FN
    pred_positives = K.sum(pred, axis=0)         # = TP + FP
    true_positives = K.sum(true * pred, axis=0)  # = TP
        #all with shape (4,)

    precision = (true_positives + K.epsilon()) / (pred_positives + K.epsilon()) 
    recall = (true_positives + K.epsilon()) / (ground_positives + K.epsilon()) 
        #both = 1 if ground_positives == 0 or pred_positives == 0
        #shape (4,)

    f1 = 2 * (precision * recall) / (precision + recall + K.epsilon())
        #not sure if this last epsilon is necessary
        #matematically not, but maybe to avoid computational instability
        #still with shape (4,)

    weighted_f1 = f1 * ground_positives / K.sum(ground_positives)
    weighted_f1 = K.sum(weighted_f1)

    return 1 - weighted_f1 #for metrics, return only 'weighted_f1'
```

When we train a Neural Network we must first define some essential values that are going to determine how the Neural Networks is going to learn from the data that we feed it with, the values that we are going to define here are not all of the values/parameters that Neural Networks used and need to learn better, the parameters that we are going to manually defined are:
<ul>
    <li><code>Epochs</code>: This define how many times the Neural Network will be train from the data.</li>
    <li><code>Batch Size</code>: It is the quantity of samples that are going to be taken from the data, this samples will be 
        feedforward into the Neural Networks and the Neural Network will learn from this batch and be evaluate from it.</li>
    <li><code>Learning Rate</code>: Is the factor that it is used to determine the step size of an optimization algorithm at 
        each iteration while moving toward a minimum of a loss function.</li>
    <li><code>Hidden Neurons</code>: These compose the layer that will do math functions in order to infere from the data that 
        is fed into them, these neurons are nodes, what we are defined is the number of nodes that the hidden layers are going 
        to have.</li>
</ul>

```python
EPOCHS = 48
BATCH_SIZE = 16
LR_RATE = 0.001
HIDDEN_NEURONS = 64
```

In the method <code>create_model()</code>, we defined the architecture of the <b>Multilayer Perceptron</b>  that we are going to use, this architechture is define by:
<ul>
    <li>The Input layer, takes the number of columns from the training set and outputs the number of 
        <code>HIDDEN_NEURONS</code>, what it outputs are weights, these weights are calculated by applying a kernel and an 
        activation function.</li>
    <li>A Hidden layer, that takes as input the number of nodes from the last layer, and it outputs <i>new</i> weights that are 
        calculated by applying a kernel and an activation function and these weights are passed to another layer.</li>
    <li>The Output layer, once again the layer takes weights from the last layer and processes them, as this is the output 
        layer, it uses another activation function that calculates the probability of a sample belonging to a class.</li>
</ul>
Then inside of this method, we specify the optimizer algorithm that will optimized the weights and try to reduce the <b>loss function</b>, we are using <code>Adam</code> optimizer, with a <code>learning_rate</code> of <b>0.001</b>, what this means is that the result from the optimizer will be multiplicated by the <codE>learning_rate</code>, for more information of the parameters that Adams has we can check the <a href='https://keras.io/optimizers/#adam'>official documentation</a> of <code>Keras</code> and for knowing how it works we can take a read at this <a href='https://towardsdatascience.com/understanding-adam-how-loss-functions-are-minimized-3a75d36ebdfc'>post</a>.
<br>
<br>
Then we compile the model, with the method <code>compile</code>, here we chose the <code>loss</code> function as <code>sparse_categorical_crossentropy</code>, this function will be the one that the optimizer will try to reduce, we also specify the optimizer that we are using and the metrics that will measure how well the Neural Network might be doing.


```python
# define a function to build the keras model
def create_model(hidden_neurons=HIDDEN_NEURONS, lr_rate=LR_RATE):
    
    # Create a Sequential model
    model = Sequential()
    # this helps mimic noise or missing data
    model.add(Dense(hidden_neurons, input_dim=input_dim(), kernel_initializer='normal', activation='relu'))
    model.add(Dense((hidden_neurons/2), kernel_initializer='normal', activation='relu'))
    model.add(Dense(2, activation='sigmoid'))
    
    # compile model
    adam = Adam(learning_rate=lr_rate)
    model.compile(loss='sparse_categorical_crossentropy', optimizer=adam, metrics=['accuracy', f1_weighted])
    return model
```

We create a Neural Network by calling the method <code>create_model()</code>, the method will return a <b>Machine Learning Perceptron</b> model, this model is keep in the object <code>model</code>, then with the method <code>summary</code> we display information related to the architecture of the Neural Network we design, in this case it has an architecture of a <b>Machine Learning Perceptron</b>, this is a very simple architecture <i> but is it powerful?</i> We will see. 

```python
model = create_model()
model.summary()
```

At this time we have defined the Neural Network that we are going to use to learn from the ASD dataset, what we are going to do next is to <b>train</b> the Neural Network, this is done by calling the method <code>fit</code> from the object <code>model</code>, we have to pass the training sets, number of epochs, the size of the batch and the proportion of data from the train set to use as a validation set, (<i>having a validation set</i> helps to noticed how well the Multi Layer Perceptron learned during the training phase), these are the required parameters to start training the Neural Network.
<br>
<br>
<b>Note</b>: There are more parameters that can be specified to improve the performance of the Neural Network and reduce the training time, for reducing the training time we can set the parameter <code>use_multiprocessing</code> to <code>True</code>. To learn more about the parameter that the method <code>fit()</code> has we can check this <a href='https://keras.io/models/model/#fit'>link</a> from the official documentation of <code>Keras</code>.

```python
model_history = model.fit(X_train,
                          y_train, 
                          epochs=EPOCHS, 
                          batch_size=BATCH_SIZE, 
                          verbose=2, 
                          validation_split=0.01,
                          validation_data=None,
                          shuffle=True,
                          class_weight=None,
                          sample_weight=None,
                          initial_epoch=0,
                          steps_per_epoch=None,
                          validation_steps=None,
                          validation_freq=1,
                          max_queue_size=10,
                          workers=1,
                          use_multiprocessing=True)
```

Cool! The <b>MLP</b> was trained, let's plot the result of <code>loss</code>, <code>validation loss</code>, <code>accuracy</code>, <code>validation accuracy</code>, <code>f1 weighted</code> and <code>validation f1 weighted</code>  from the training:

```python
def plot_accuracies_losses(title, xlabel, ylabel,
                           train_loss=model_history.history["loss"],
                           val_loss=model_history.history["val_loss"],
                           accuracy=model_history.history["accuracy"],
                           val_accuracy=model_history.history["val_accuracy"],
                           f1_weighted=model_history.history['f1_weighted'],
                           val_f1_weighted=model_history.history['val_f1_weighted'],
                           epochs=EPOCHS):
    fig = go.Figure()
    x_range = np.arange(1, epochs+1)
    
    fig.add_trace(go.Scatter(
        y=val_loss,
        x=x_range,
        name="Val Loss",
        mode='lines+markers'
    ))

    fig.add_trace(go.Scatter(
        y=train_loss,
        x=x_range,
        name="Train Loss",
        mode='lines+markers'
    ))
    
    fig.add_trace(go.Scatter(
        y=accuracy,
        x=x_range,
        name="Accuracy",
        mode='lines+markers'
    ))
    
    fig.add_trace(go.Scatter(
        y=val_accuracy,
        x=x_range,
        name="Val Accuracy",
        mode='lines+markers'
    ))
    
    fig.add_trace(go.Scatter(
        y=f1_weighted,
        x=x_range,
        name="F1 Weighted",
        mode='lines+markers'
    ))
    
    fig.add_trace(go.Scatter(
        y=val_f1_weighted,
        x=x_range,
        name="Val F1 Weighted",
        mode='lines+markers'
    ))

    fig.update_layout(
        title=title,
        xaxis_title=xlabel,
        yaxis_title=ylabel,
        font=dict(
            family="Courier New, monospace",
            size=18,
            color="#7f7f7f"
        )
    )

    fig.show()
```

```python
plot_accuracies_losses("Training Losses and Metrics on Autism Spectrum Disorder Dataset", "Epochs", "Losses and Metrics")
```

The <b>MLP</b> model did well during the training phase, now let's predict some samples from the test set.

The result of the metrics <code>accuracy</code> and <code>f1 weighted score</code>, are not bad, they are good, these results are from the <b>MLP</b> base model, <i>the result from the <code>classification_report</code> are rounded</i>, let's see the unrounded score of the <code>f1 weighted score</code>

<h2>Hyperparameter Tuninig for a Machine Learning Perceptron Neural Network Architecture</h2>
<br>
When it is the time to beat the performance of a base model, we have several techniques to do it, all of these techniques have their pros and cons. These techniques do not asure that they are going to find the set of hyperparameters that are going to help the new model to achieve a better performance than the base model. We are going to do the Hyperparameter tuning in a single parameter, the <code>learning rate</code>, our intentions are to find the best <code>learning rate</code> for the current configuration of our actual <b>MLP</b> architecture.
<br>
<br>
Hyperparameter Tuning for Neural Networks is a very difficult task due to the high computational resources that needs to be properly done, a Neural Network usually have more parameter for tuning than standard machine learning algorithms, if we were interested to properly tuned the paramater of the <b>MLP</b> architecture we built, we should have to tune these parameters:
<ul>
    <li><code>Model Architecture <i>Dense Layers</i></code> <i>for each dense layer the hyperparameters could be different</i>:

Lots of hyperparameters! It will be very computationally expensive to properly applied hyperparameter tuning for just a simple <b>MLP</b> architecture, that is the main reason why I am only focusing hyperparameter tuning for the <code>learning rate</code> parameter, the other issue of Hyperparameter Tuning is <i><b>defining the domain space</b></i>, the domain space is the range of parameters that are going to be tested.
<br>
<br>
As we mentioned earlier, there are several techniques to apply Hyperparameter Tuning, they are:
<ul>
    <li><code>Bayesian Optimization</code>, test the hyperparameters according bayesian probability, it is usually slow and 
        does not test the whole domain space of hyperparameters <i>but it can return the best set of hyperparameters.</i></li>
    <li><code>Baby Sitting</code>, <i>it is considered a bad practice even though is a good technique to define a domain 
        space</i></li>
    <li><code>Random Search Cross Validation</code>, it saves time but does not test the whole domain space of hyperparameters, 
        therefore we will not know if it obtained the best set of hyperparameters.</li>
        <li>*<code>Grid Search Cross Validation</code>, <i>we are going to use this technique but with <code>KFold</code> 
            validation.</i> It is usually the slowest from the four techniques but it always returns the best set of 
            hyperparameters from the domain space.</li>
</ul>

Let's define the domain space for the <code>learning rate</code> hyperparameter, this domain space is defined with the <code>arange()</code> function from the <code>numpy</code> package, we are going to create a dictionary called <code>params</code>, in this dictionary you can add as much paramaters as you wish, just have in mind the structure of this dictionary: <i>the <code>key</code> will be the name of the hyperparameter and the <code>value</code> will be the domain space of hyperparameters that will be tested.</i>

```python
params = dict(lr_rate=np.arange(0.001, 1.1, 0.00125))
```

The next fucntion returns the number of sets that are going to be tested in order to obtained the best model built with the best set of parameters

```python
def quantity_of_models(params=params):
    result=1
    for key in params:
        result = result * len(params[key])

    return result 
```

Let's execute the function:

```python
quantity_of_models()
```

In this case we are trying to get a better <code>learning rate</code>, we are going to test <b>880</b> learning rates, <i>that it is a lot of models</i> and remember we are using <code>KFold</code>, this means that we are going to test <b>$880 * n_{folds}$</b> models. Let's use the class <code>ParameterGrid()</code> from <code>sklearn</code> this class allows us to create an object of parameters for testing, we will call this object <code>params</code>

```python
params = ParameterGrid(params)
```

The cell bellow is really simple, this is what it does:
<ol>
    <li>We create a list called <code>model_params</code>, here we are going to keep the parameters tested.</li>
    <li>We create a list called <code>f1_weighted_scores</code>, here we are going to keep the <code>f1 weighted score</code>
        from the tested model.</li>
    <li>We create a list called <code>accuracy_scores</code>, here we are going to keep the <code>accuracy score</code>
        from the tested model.</li>
    <li>We create <code>numpy arrays</code> from the train sets, we called them <code>X_kfold</code> and <code>y_kfold</code>.
    </li>
    <li>Then we create a <code>KFold</code> object, this object will help us in performing cross validation to test the 
        parameters, we will call it <code>kf</code>, I will use <b>10</b> splits, feel free to set how many splits you consider 
        adequate.</li>
    <li>We enter in a <code>for</code> loop, that will iterate through the parameter grid, called <code>params</code>:
        <ol>
            <li>We create the <code>lr_rate</code> variable that will hold the value from <code>p['lr_rate']</code>, <i>if 
                there are more parameters to test, create a variable for every of them and assign to the variables the value 
                from the iteratior <code>p</code>, remember <code>p</code> works as a dictionary, to get the value for the 
                specific parameter, you only need to specify the key, example: <code>p['key']</code>.</i></li>
            <li>We create a varibale to keep the <code>f1 weighted score</code>, we call it <code>f1_weighted_score</code>.
            </li>
            <li>We create a varibale to keep the <code>accuracy score</code>, we call it <code>accuracy</code>
            </li>
            <li>We enter to another <code>for</code> loop, this loop is for applying <code>KFold</code> validation, we create
                two iterators <code>train_index</code> and <code>test_index</code>, the object that we will be iterating 
                through is the result from <code>kf.split(X_kfold)</code>
            <ol>
                <li>We first create the training set, here we defined two arrays <code>X_train_kfold</code> and 
                    <code>X_test_kfold</code> they will have the result from <code>X_kfold[train_index]</code> and 
                    <code>X_kfold[test_index]</code>.</li>
                <li>Then we create the test set, here we defined two arrays <code>y_train_kfold</code> and 
                    <code>y_test_kfold</code> they will have the result from <code>y_kfold[train_index]</code> and 
                    <code>y[test_index]</code>.</li>
                <li>We create the model, by calling the method <code>create_model()</code>, we pass the parameter 
                    <code>lr_rate</code>, we store the returned model in the object <code>model</code>.</li>
                <li>It is time to start the training phase the model, for this we create a history object call 
                    <code>model_history</code>, this object will have the information of the metrics and losses that we 
                    required to see how well or bad the model with the learning rate is doing.</li>
                <li>With <code>y_pred_kfold</code> we save the predictions from calling }
                    <code>model.predict_classes(<i>X_test_kfold, batch_size=BATCH_SIZE</i>)</code>
                    .</li>
                <li>In <code>f1_weighted_score<i>+=</i></code> we calculate the <code>f1 weighted score</code> from the model 
                    predictions with <code>f1_score(<i>y_test_kfold, y_pred_kfold, average='weighted'</i>)</code>.</li>
                <li>Now in <code>accuracy<i>+=</i></code> we calculate the <code>accuracy score</code> from the model 
                    predictions with <code>accuracy_score(<i>y_test_kfold, y_pred_kfold</i>)</code>.</li>
            </ol>
        </ol>
    </li>
    <li>Now in <code>f1_weighted_score</code> we do the next calculation <code>f1_weighted_score/<i>n<sub>folds</sub></i></code>, this is 
        for calculating the average <code>f1_weighted_score</code> from the $n_i$ fold.</li>
    <li>Now in <code>accuracy</code> we do the next calculation <code>accuracy/<i>n<sub>folds</sub></i></code>, this is 
        for calculating the average <code>accuracy</code> from the $n_i$ fold.</li>
    <li>We append in <code>f1_weighted_scores</code> the results from <code>f1_weighted_score</code>.</li>
    <li>We append in <code>accuracy_scores</code> the results from <code>accuracy</code>.</li>
    <li>We append in <code>model_params</code> the parameters from <code>p</code>.</li>
</ol>

```python
model_params = list()
f1_weighted_scores = list()
accuracy_scores = list()

X_kfold = np.array(X_train)
y_kfold = np.array(y_train)

kf = KFold(n_splits=10, random_state=seed_value, shuffle=False)

for p in params:
    
    lr_rate = p['lr_rate']
    
    f1_weighted_score = 0
    accuracy = 0
    
    for train_index, test_index in kf.split(X_kfold):
    
        X_train_kfold, X_test_kfold = X_kfold[train_index], X_kfold[test_index]
        y_train_kfold, y_test_kfold = y_kfold[train_index], y_kfold[test_index]

        model = create_model(hidden_neurons=HIDDEN_NEURONS, lr_rate=lr_rate)

        model_history = model.fit(X_train_kfold,
              y_train_kfold, 
              epochs=EPOCHS, 
              batch_size=BATCH_SIZE, 
              verbose=0, 
              validation_split=0.01,
              validation_data=None,
              shuffle=True,
              class_weight=None,
              sample_weight=None,
              initial_epoch=0,
              steps_per_epoch=None,
              validation_steps=None,
              validation_freq=1,
              max_queue_size=10,
              workers=1,
              use_multiprocessing=True)

        y_pred_kfold = model.predict_classes(X_test_kfold, batch_size=BATCH_SIZE)
        
        f1_weighted_score+= f1_score(y_test_kfold, y_pred_kfold, average='weighted')
        accuracy+= accuracy_score(y_test_kfold, y_pred_kfold)
    
    f1_weighted_score = f1_weighted_score/10
    accuracy = accuracy/10
    
    f1_weighted_scores.append(f1_weighted_score)
    accuracy_scores.append(accuracy)
    
    model_params.append(p)    
```

In <code>loo</code> we create the object <code>LeaveOneOut()</code>, then with <code>loo.get_n_splits(<i>X</i>)</code> we can know how many splits we are going to do, <i>these splits means how many times we are going to train and test the model with the best learning rate.</i>
